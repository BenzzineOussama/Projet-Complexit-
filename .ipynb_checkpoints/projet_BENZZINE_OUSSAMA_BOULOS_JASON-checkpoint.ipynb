# **Théorie de la Complexité des Algorithmes**

## **Introduction**
La complexité d'un algorithme est une mesure de l'efficacité de celui-ci en termes de **temps d'exécution** ou d'**espace mémoire** requis, en fonction de la taille de l'entrée. Cette évaluation permet de comparer différents algorithmes pour résoudre un même problème.

---

## **Types de Complexités**
- **Complexité temporelle** (*Time Complexity*) : Mesure le nombre d’opérations élémentaires exécutées en fonction de la taille de l’entrée.
- **Complexité spatiale** (*Space Complexity*) : Mesure la quantité de mémoire supplémentaire utilisée par l’algorithme.

---

## **Notations Asymptotiques**
Les notations asymptotiques décrivent la croissance de la complexité pour de grandes tailles d'entrée ($n \rightarrow \infty$).

- **O (notation Big-O)** : Représente une borne supérieure. Exemple : $O(n^2)$ signifie que l'algorithme peut être aussi lent que $n^2$ opérations dans le pire cas.
- **Ω (Omega)** : Représente une borne inférieure. Exemple : $Ω(n)$ signifie qu'au minimum $n$ opérations seront nécessaires.
- **Θ (Theta)** : Représente une borne exacte. Exemple : $Θ(n^2)$ indique que $n^2$ opérations sont effectuées dans le pire et le meilleur cas.

---

## **Cas de Complexité**
Les algorithmes sont analysés selon trois cas :
- **Meilleur cas** (*Best Case*) : La performance optimale, souvent rare. Exemple : tableau déjà trié pour **Insertion Sort**.
- **Pire cas** (*Worst Case*) : Le scénario le plus défavorable. Exemple : tableau trié en ordre décroissant pour **Bubble Sort**.
- **Cas moyen** (*Average Case*) : Un cas typique basé sur une entrée aléatoire.

---

## **Complexité des Algorithmes de Tri**
Voici un résumé théorique des complexités des algorithmes de tri étudiés dans le projet :

| Algorithme          | Meilleur Cas   | Pire Cas        | Cas Moyen       | Complexité spatiale |
|---------------------|----------------|-----------------|-----------------|-------------------------|
| **Bubble Sort**     | $O(n)$         | $O(n^2)$        | $O(n^2)$        | $O(1)$                 |
| **Selection Sort**  | $O(n^2)$       | $O(n^2)$        | $O(n^2)$        | $O(1)$                 |
| **Insertion Sort**  | $O(n)$         | $O(n^2)$        | $O(n^2)$        | $O(1)$                 |
| **Heap Sort**       | $O(n\log n)$   | $O(n\log n)$    | $O(n\log n)$    | $O(1)$                 |
| **Quick Sort**      | $O(n\log n)$   | $O(n^2)$        | $O(n\log n)$    | $O(\log n)$            |
| **Merge Sort**      | $O(n\log n)$   | $O(n\log n)$    | $O(n\log n)$    | $O(n)$                 |

---

## **Importance de l'Évaluation Empirique**
- Les complexités théoriques ne prennent pas en compte les optimisations pratiques, comme les caches ou les architectures matérielles.
- Les tests empiriques complètent les analyses théoriques en illustrant les performances sur des cas concrets.

---

## **Applications Pratiques**
- Choisir l'algorithme adapté au contexte (ex : Quick Sort est rapide pour des petits tableaux, Merge Sort est stable et efficace pour des grandes données avec des besoins de stabilité).

## **Bubble Sort**

### **Description de l'algorithme**

Le Bubble Sort compare les éléments adjacents d'un tableau et les échange si nécessaire, jusqu'à ce que le tableau soit trié. Le processus est répété pour tous les éléments non triés.

---

### **Analyse théorique**

#### **Pire cas (\($O(n^2)$\))** :
- Lorsqu'un tableau est trié dans l'ordre inverse.
- Chaque élément doit être comparé avec tous les autres, ce qui donne un nombre total de comparaisons de :
  $$
  (n-1) + (n-2) + \dots + 1 = \frac{n(n-1)}{2}
  $$
  soit \($O(n^2)$\).

#### **Meilleur cas (\($O(n)$\))** :
- Lorsque le tableau est déjà trié.
- Une seule passe est nécessaire pour vérifier que tous les éléments sont à leur place.

#### **Cas moyen (\($O(n^2)$\))** :
- En supposant des données aléatoires, chaque élément peut nécessiter d’être échangé plusieurs fois, mais le nombre total de comparaisons reste proportionnel à \(O(n^2)\).

#### **Complexité spatiale** :
- **En place** : Utilise \($O(1)$\) mémoire supplémentaire.
- **Stable** : Oui, car les éléments égaux conservent leur ordre initial.


### **Code Python : Implémentation de Bubble Sort**

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # Échange
                swapped = True
        # Si aucun échange n'a eu lieu, le tableau est trié
        if not swapped:
            break
    return arr
```

### **Validation empirique**

Nous allons vérifier les temps d'exécution du Bubble Sort sur des tableaux :

1. **Déjà triés (*Meilleur Cas*)**.
2. **Triés dans l'ordre décroissant (*Pire Cas*)**.
3. **Avec des valeurs aléatoires (*Cas Moyen*)**.


```python
executerTri(bubble_sort,'r','Bubble sort')
```

### **Analyse du graphique**

#### **Meilleur Cas (\($O(n)$\))** :
- La courbe doit croître de manière **linéaire**. 
- Ici, la courbe en tirets semble correcte, car elle reste en bas et augmente proportionnellement à \(n\).

#### **Cas Moyen (\($O(n^2)$\))** :
- La courbe doit avoir une croissance **quadratique**. 
- La ligne continue reflète bien ce comportement.

#### **Pire Cas (\($O(n^2)$\))** :
- Comme pour le cas moyen, la croissance doit être **quadratique**, mais un peu plus élevée car le pire cas nécessite plus d'opérations.
- La courbe en pointillés semble bien capturer cela.

### **Conclusion**

Les résultats empiriques obtenus dans le graphique sont cohérents avec les analyses théoriques des complexités des cas :

1. **Meilleur Cas (\($O(n)$\))** :
   - Le temps d'exécution est linéaire lorsque les données sont déjà triées, comme prévu théoriquement.
   - La courbe en tirets illustre cette performance optimale.

2. **Cas Moyen (\($O(n^2)$\))** :
   - La courbe continue suit une tendance quadratique, ce qui reflète bien la complexité théorique dans un scénario aléatoire.

3. **Pire Cas (\($O(n^2)$\))** :
   - La courbe en pointillés montre une croissance quadratique plus élevée, en accord avec le pire cas où le nombre maximum d'opérations est requis.

### **Interprétation générale** :
- Le **Bubble Sort** est un algorithme simple mais inefficace pour des grands tableaux, en raison de sa complexité quadratique (\($O(n^2)$\)) dans la plupart des cas.
- Il est performant uniquement pour des tableaux presque triés ou de petite taille.

## **Merge Sort**

### **Description de l'algorithme**

Merge Sort divise récursivement le tableau en deux moitiés, trie chaque moitié, puis fusionne les sous-tableaux triés pour produire le tableau final trié.

---

### **Analyse théorique**

#### **Complexité temporelle (\($O(n \log n)$\))** :
La complexité est la même pour les trois cas (meilleur, pire et moyen), car l'algorithme effectue toujours les mêmes étapes, indépendamment de l'ordre initial des éléments.

1. **Division** :
   - Le tableau est divisé récursivement en deux moitiés.
   - Cela génère \(\log n\) niveaux, car chaque division réduit la taille du tableau de moitié.

2. **Fusion** :
   - À chaque niveau, les sous-tableaux sont fusionnés, ce qui nécessite \(O(n)\) opérations pour comparer et réorganiser tous les éléments.

3. **Combinaison des deux** :
   - Le coût total est \(O(n)\) par niveau de fusion multiplié par \(log n\) niveaux, soit :
   $$
   T(n) = O(n \cdot \log n)
   $$

**Conclusion** : Merge Sort a une complexité temporelle uniforme (\(O(nlog n)\)) pour tous les cas, car il suit les mêmes étapes systématiquement.

---

#### **Complexité spatiale (\($O(n)$\))** :
- Merge Sort utilise \(O(n)\) de mémoire supplémentaire pour stocker les sous-tableaux temporaires nécessaires pendant la fusion.
- L'algorithme est **stable**, car les éléments égaux conservent leur ordre initial dans le tableau trié.

---

#### **Pourquoi la complexité est identique pour tous les cas ?**
Contrairement à certains algorithmes comme Quick Sort, Merge Sort suit une structure fixe :
- Il divise toujours le tableau complètement, peu importe l'ordre des éléments.
- Les étapes de fusion nécessitent systématiquement \(O(n)\), indépendamment de l'état initial des données.
